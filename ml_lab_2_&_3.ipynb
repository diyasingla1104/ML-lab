```python
# Import necessary libraries
import pandas as pd

# Upload the file
from google.colab import files
uploaded = files.upload()  # Prompts you to upload your Excel file

# Load the Excel file into a Pandas DataFrame
file_name = list(uploaded.keys())[0]  # Gets the uploaded file name
df = pd.read_excel(file_name)

# Display the first few rows to understand the structure
print("DataFrame Head:")
print(df.head())

# Show available columns
print("\nAvailable Columns in the DataFrame:")
print(df.columns.tolist())

# Function for extracting a specific column
def extract_column():
    column_name = input("\nEnter the column name you want to extract: ").strip()
    if column_name in df.columns:
        print(f"\nData in column '{column_name}':")
        print(df[column_name])
    else:
        print(f"\nColumn '{column_name}' not found!")

# Function for extracting a specific row
def extract_row():
    index_value = input("\nEnter the index (row number) you want to extract: ").strip()
    try:
        index_value = int(index_value)  # Convert to integer
        if 0 <= index_value < len(df):
            print(f"\nData in row {index_value}:")
            print(df.iloc[index_value])
        else:
            print("\nInvalid row number!")
    except ValueError:
        print("\nPlease enter a valid integer for the row index!")

# Function to search within rows
def search_row():
    search_column = input("\nEnter the column name to search in: ").strip()
    if search_column in df.columns:
        search_value = input(f"Enter the value to search for in column '{search_column}': ").strip()
        result = df[df[search_column] == search_value]
        if not result.empty:
            print(f"\nRows matching '{search_value}' in column '{search_column}':")
            print(result)
        else:
            print(f"\nNo rows found with '{search_value}' in column '{search_column}'!")
    else:
        print(f"\nColumn '{search_column}' not found!")

# Menu for user interaction
while True:
    print("\nOptions:")
    print("1. Extract a specific column")
    print("2. Extract a specific row")
    print("3. Search for data within rows")
    print("4. Exit")

    choice = input("\nEnter your choice (1-4): ").strip()
    if choice == "1":
        extract_column()
    elif choice == "2":
        extract_row()
    elif choice == "3":
        search_row()
    elif choice == "4":
        print("\nExiting the program.")
        break
    else:
        print("\nInvalid choice! Please enter a number between 1 and 4.")
```
